```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_csv("hiring.csv")
```

```python
df
```

```python
df.experience = df.experience.fillna('zero')
```

```python
df.experience
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
    
```

```python
df.head()
```

```python
df.columns
```

```python
df["test_score(out of 10)"] = df["test_score(out of 10)"].fillna(df["test_score(out of 10)"].median())
```

```python
df
```

```python
df.isna().sum()
```

```python
mlv_reg = linear_model.LinearRegression()
```

```python
mlv_reg.fit(df.drop('salary($)',axis='columns'),df['salary($)'])
```

```python


# 12 yr experience, 10 test score, 10 interview score

print(mlv_reg.predict([[2,9,6]])) # 2 yr experience, 9 test score, 6 interview score


print(mlv_reg.predict([[12,10,10]]))
```

```python
mlv_reg.intercept_
```
