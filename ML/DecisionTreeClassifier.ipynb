```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
data = pd.read_csv("titanic.csv")
data.head()

data_new = data[["Pclass","Age","Sex","Fare","Survived"]]
data_new.head()
```

```python
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.preprocessing import LabelEncoder
dfle = LabelEncoder()
data_new.Sex = dfle.fit_transform(data_new.Sex)
data_new.Sex
```

```python
data_new.head()
```

```python
inputs = data_new.drop("Survived",axis="columns")
target = data_new.Survived
model.fit(inputs,target)
```

```python
model
```

```python
model.predict([[1,38,0,71]])
```

```python
model.score(inputs,target)
```
