```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df1)
```

```python
df1.MultipleLines = df1.MultipleLines.replace("No phone service",0)
```

```python
df1.MultipleLines.unique()
```

```python
yes_no_cols =['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_cols:
    df1[col].replace({'Yes': 1,'No': 0,"No internet service":0},inplace=True)
```

```python
df1.dtypes
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1.columns
```

```python
for col in df1.columns:
    print(f'{col} : {df1[col].unique()}')
```

```python
df1.gender.unique()
```

```python
#one hot encoding for columns such as 'InternetService', 'PaymentMethod', 'Contract' .

df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
for col in df2.columns:
    df2[col].replace({False:0,True:1},inplace=True)
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
mc_churn_no = df1[df1.Churn==0].MonthlyCharges      
mc_churn_yes = df1[df1.Churn==1].MonthlyCharges      

plt.xlabel("Monthly Charges")
plt.ylabel("Number Of Customers")
plt.title("Customer Churn Prediction Visualiztion")


plt.hist([mc_churn_yes, mc_churn_no], rwidth=0.95, color=['green','red'],label=['Churn=Yes','Churn=No'])
plt.legend()
```

```python
X = df2.drop("Churn",axis = 'columns')
Y = df2.Churn
```

```python
df2.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2 , random_state = 5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp = model.predict(x_test)
```

```python
yp[:5]
```

```python
y_pred =[]
for ele in yp:
    if ele>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
